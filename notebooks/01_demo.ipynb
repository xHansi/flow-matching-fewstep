# Demo — Few-Step Generation with Flow Matching

A minimal end-to-end demo on a Kaggle T4: train a small Flow Matching model on MNIST and
show that it generates recognizable digits in **very few sampling steps**. For the full
study (DDPM baseline, reflow, FID sweep, domain shift) see `02_results.ipynb`; for a
narrated visual tour see `03_walkthrough.ipynb`.

## Train a small Flow Matching model

A short run for the demo (5 epochs). Increase `--epochs` for sharper samples.

```python
!cd {REPO} && PYTHONPATH={REPO} python -m scripts.train --method flow --dataset mnist --epochs 5 --preview-steps 8 --preview-cfg 2.0
```

## The point: few-step sampling

The same trained model, sampled with 1, 2, 4 and 8 Euler steps. Digits are already
recognizable at 1–2 steps — that is the property this project studies.

```python
!cd {REPO} && PYTHONPATH={REPO} python -m scripts.sample --ckpt results/flow_mnist/ckpt.pt --steps 1,2,4,8 --per-class 8 --cfg-scale 2.0
```

```python
for s in (1, 2, 4, 8):
    print(f'{s} step(s)'); display(Image(f'{REPO}/results/flow_mnist/samples_steps{s}.png'))
```
